```python
# crud operation on student data 

students = {}

def create():
    sid = input("ID: ")
    if sid in students:
        print("Exists!")
    else:
        name = input("Name: ")
        age = input("Age: ")
        students[sid] = {"name": name, "age": age}
        print("Added.")

def read():
    sid = input("ID: ")
    if sid in students:
        s = students[sid]
        print(f"{sid}: {s['name']}, {s['age']}")
    else:
        print("Not found.")

def update():
    sid = input("ID: ")
    if sid in students:
        name = input("New name (blank=skip): ")
        age = input("New age (blank=skip): ")
        if name: students[sid]['name'] = name
        if age: students[sid]['age'] = age
        print("Updated.")
    else:
        print("Not found.")

def delete():
    sid = input("ID: ")
    if sid in students:
        del students[sid]
        print("Deleted.")
    else:
        print("Not found.")

def show():
    if students:
        for sid, s in students.items():
            print(f"{sid}: {s['name']}, {s['age']}")
    else:
        print("Empty.")
while True:
    print("\n1.Create 2.Read 3.Update 4.Delete 5.Show 6.Exit")
    ch = int(input("Choice: "))
    if ch == 1:
        create()
    elif ch == 2:
        read()
    elif ch == 3:
        update()
    elif ch == 4:
        delete()
    elif ch == 5:
        show()
    elif ch == 6:
        break
    else:
        print("Invalid")
```

```python
students = []   # list of student dictionaries 
def create():
    sid = input("ID: ")
    for s in students:
        if s["id"] == sid:
            print("Exists!")
            return
    name = input("Name: ")
    age = input("Age: ")
    students.append({"id": sid, "name": name, "age": age})
    print("Added.")

def read():
    sid = input("ID: ")
    for s in students:
        if s["id"] == sid:
            print(f"{sid}: {s['name']}, {s['age']}")
            return
    print("Not found.")

def update():
    sid = input("ID: ")
    for s in students:
        if s["id"] == sid:
            name = input("New name (blank=skip): ")
            age = input("New age (blank=skip): ")
            if name: s["name"] = name
            if age: s["age"] = age
            print("Updated.")
            return
    print("Not found.")

def delete():
    sid = input("ID: ")
    for s in students:
        if s["id"] == sid:
            students.remove(s)
            print("Deleted.")
            return
    print("Not found.")

def show():
    if students:
        for s in students:
            print(f"{s['id']}: {s['name']}, {s['age']}")
    else:
        print("Empty.")

while True:
    print("\n1.Create 2.Read 3.Update 4.Delete 5.Show 6.Exit")
    ch = int(input("Choice: "))
    if ch == 1:
        create()
    elif ch == 2:
        read()
    elif ch == 3:
        update()
    elif ch == 4:
        delete()
    elif ch == 5:
        show()
    elif ch == 6:
        break
    else:
        print("Invalid")
print(students)
```
