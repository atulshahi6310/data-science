# genrate random colore for five diff operation using function

```python
import random

def generate_random_color():
  r = random.randint(0, 255)
  g = random.randint(0, 255)
  b = random.randint(0, 255)
  return (r, g, b)

print("Random colors :")
for i in range(5):
  color = generate_random_color()
  print(f"RGB{color}")
```

## find sum of all prime number in given range


```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def sum_of_primes(start, end):
    return sum(num for num in range(start, end + 1) if is_prime(num))

start = int (input("enter the start number "))
end = int (input("enter  the end  index"))
print("Sum of primes:", sum_of_primes(start, end))
```
