```python
class Student:
    def __init__(self, name, marks):
        self.__marks = marks      # private
        self.name = name

    def get_marks(self):          # getter
        return self.__marks

    def set_marks(self, marks):   # setter
        if marks >= 0:
            self.__marks = marks
        else:
            print("Invalid marks")
```

```python
st1 = Student("Atul" , "90%")
```

```python
print(st1.name)
```

```python
try :
    print(st1.marks)
except Exception as e:
    print(e)
    
```

```python
st1.set_marks(99)
print(st1.get_marks())
```
